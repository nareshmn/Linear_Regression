```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("income.csv")
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('Per Capita Income')
plt.scatter(df.year,df.pci)
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df.pci)
```

```python
reg.predict([[2050]])
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('Percapita Income')
plt.scatter(df.year,df.pci,color="red",marker='+')
plt.plot(df.year,reg.predict(df[['year']]),color='blue')
```

```python
df1=pd.read_csv("incomepredict.csv")
df1.head()
```

```python
p=reg.predict(df1[['year']])
p
```

```python
df1['pci']=p
df1
```

```python
df1.to_csv("predictedincome.csv",index=False)
```
